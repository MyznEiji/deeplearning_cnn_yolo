```python
from keras.datasets import cifar10
```

```python
import matplotlib.pyplot as plt
```

```python

# Divide into training data adn test data
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Define the list of label
cifar10_labels = [
    'airplane',
    'automobile',
    'bird',
    'cat',
    'deer',
    'dog',
    'frog',
    'horse',
    'ship',
    'truck'
]
```

```python
# Description
for index, img in enumerate(X_train[:30]):
    plt.subplot(3, 10, index + 1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(cifar10_labels[y_train[index][0]])

plt.show()
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255.
X_test /= 255.
```

```python
from keras.utils import np_utils

# Convert class label to one-hot-vector
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPool2D

from keras.layers.core import Dropout

# Initialize the model
model=Sequential()

# Design the layer =============================
# Initialize the model
model=Sequential()

# Design the layer =============================
## Add convolution layer (Conv2D)
model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(32,32,3)))
model.add(Conv2D(32, kernel_size=(3,3), activation='relu'))

## Add pooling layer 
model.add(MaxPool2D(pool_size=(2,2)))

## Well convolution =>Add drop layer as overfitting measures after pooling was run one time
model.add(Dropout(0.25))

### Convolution again=> Repeat pooling
model.add(Conv2D(64, kernel_size=(3,3), activation='relu', padding='same'))
model.add(Conv2D(64, kernel_size=(3,3), activation='relu', padding='same'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))

## Total binding layer in the last
model.add(Flatten()) # flattening
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10,activation='softmax'))       # Last activation function is softmax function
# ===================================
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(
    X_train,
    y_train,
    batch_size=32,
    epochs=20,
    validation_split=0.2
)
```
